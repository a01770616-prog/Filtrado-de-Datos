```python
#Importamos librer√≠as requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

CARGA DE ARCHIVO XLSX

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_csv('Datos_FacturacioÃÅn.csv')
df
```

```python
#Verificamos informaci√≥n del DataFrame
df.info()
```

**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
# 1-CVE_CLPV      (valores: 1000 a 2000)
#Filtro Y
filtro1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1
# Las variables filtradas fueron: La √∫nica variable (columna) utilizada para filtrar fue: CVE_CLPV. Y los valores filtrados son los que cumplen la condici√≥n: mayores o iguales a 1000, y menores o iguales a 2000.
```

```python
df6 = filtro1.groupby(['CVE_CLPV'])['CVE_CLPV'].count()
df6
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Entregable1.csv")
```

```python
# 2-CVE_VEND     (Todas las claves excepto ‚Äú5‚Äù y ‚Äú4‚Äù)
filtro2 = df[~(df["CVE_VEND"] == 4) & ~ (df["CVE_VEND"] == 5)]
filtro2
# Las variables filtradas fueron: La variable (columna) filtrada fue: CVE_VEND. La condici√≥n aplicada fue: Excluir los registros donde CVE_VEND sea 4 o 5.
```

```python
df6 = filtro1.groupby(['CVE_CLPV'])['CVE_CLPV'].count()
df6
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Entregable2.csv")
```

```python
# 3-FECHA_ENT   (Fechas de ‚Äú28/02/2022‚Äù)
filtro3 = df[df["FECHA_ENT"] == "2022-02-28 00:00:00"]
filtro3
# Las variables filtradas fueron: La variable (columna) filtrada fue: FECHA_ENT. La condici√≥n aplicada fue: Seleccionar √∫nicamente los registros donde FECHA_ENT sea ‚Äú2022-02-28 00:00:00‚Äù (equivalente al 28/02/2022). 
```

```python
df6 = filtro2.groupby(['FECHA_ENT'])['FECHA_ENT'].count()
df6
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("Entregable3.csv")
```

```python
# 4-CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (‚ÄúE‚Äù)
filtro4 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
# Las variables filtradas fueron: Las variables (columnas) filtradas fueron: CAN_TOT (se seleccionan los registros con valores menores a 5951.7) STATUS (se seleccionan los registros cuyo valor es "E").
```

```python
df6 = filtro4.groupby(['CAN_TOT'])['CAN_TOT'].count()
df6
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("Entregable4.csv")
```

```python
# 5-Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5 = df[["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
filtro5
# Las variables filtradas fueron:Las variables (columnas) filtradas fueron: CVE_DOC, FECHA_ENT, FECHA_VEN, CAN_TOT.
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("Entregable5.csv")
```

```python
#6-Solo las filas de 7001-7099
filtro6 = df.iloc[7001:7100, :]
filtro6
# Las variables filtradas fueron: Entonces, las variables filtradas fueron: todas las columnas, y las filas filtradas fueron 7001 a 7099.
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("Entregable6.csv")
```

```python
# 7-Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)
df_indexado = df.set_index("CVE_VEND")
filtro7 = df_indexado.loc[[1, 2], ["FECHAELAB"]]
filtro7
# Las variables filtradas fueron: Las variables (columnas) filtradas fueron: FECHAELAB (columna seleccionada). El √≠ndice usado para filtrar filas fue: CVE_VEND, seleccionando los valores 1 y 2.
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Entregable7.csv")
```
